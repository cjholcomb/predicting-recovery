```python
from src.dictionaries import *
import pandas as pd
import numpy as np
```

```python
def add_qtrid(df):
    df['qtrid'] = df['year'] + (df['qtr']/4)
    return df

#import one year of data to explore
def import_one(year):
    filepath = 'data/by_county/' + str(year) + '.csv'
    df = pd.read_csv(filepath, dtype = schema_dict)
    #df = df.drop(drop_columns, axis = 1)
    return df

#import all data into a single file
def import_all(years):
    df = import_one(years[0])
    for year in years[1:]:
        df = df.append(import_one(year))
    df = add_qtrid(df)
    return df
```

```python
df = import_all(recession1_years)
```

```python
def create_df_empl(df):
    return df.pivot_table(columns = 'qtrid', values = 'month3_emplvl', index = ['area_fips', 'area_title'], aggfunc = np.sum)
```

```python
df = import_all(recession1_years)
df_empl = create_df_empl(df)
df_empl = df_empl.reset_index()
df_empl['nadir'] = df_empl.iloc[:,9:].min(axis=1)
df_empl['nadir_qtr'] = df_empl.iloc[:,9:].idxmin(axis=1)
df_empl['nadir_qtr_ct'] = (df_empl['nadir_qtr'] - 2000.25) * 4
df_empl['nadir_qtr_ct'] = ((df_empl['nadir_qtr'] - 2000) * 4).astype('int32')
df_empl['pre-peak'] = 0
df_empl['post-peak'] = 0
for index, row in df_empl.iterrows():
    slicer = row['nadir_qtr_ct'] + 3
    pre_peak = row.iloc[2:slicer].max()
    df_empl.iloc[index,37] = pre_peak
    post_peak = row.iloc[slicer:].max()
    df_empl.iloc[index,38] = post_peak
df_empl['recovery'] = df_empl['post-peak'] >= df_empl['pre-peak']
df_empl['delta'] = df_empl['post-peak'] - df_empl['pre-peak']
```
